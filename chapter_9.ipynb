#**Списки**

В этой главе рассматриваются списки, последние из трех основных линейных коллекций, обсуждаемых в этой
книге (двумя другими являются стеки и очереди). Списки поддерживают гораздо более широкий спектр операций, чем стеки и очереди, и, следовательно, являются как более широко используемыми, так и более
сложными в реализации. Хотя Python включает встроенный тип списка, существует несколько возможных реализаций, из которых Python - только одна. Чтобы разобраться в изобилии
фундаментальных операций со списком, вы классифицируете их на три группы: операции на основе индексов,
операции, основанные на содержимом, и операции, основанные на местоположении. В этой главе рассматриваются две
наиболее распространенные реализации списков: массивы и связанные структуры. Вы разрабатываете специальный
тип объекта, называемый итератором списка, для поддержки операций, основанных на местоположении.
В примере из этой главы показано, как создать список особого типа, называемый отсортированным списком.

## **Обзор списков**

Список поддерживает манипулирование элементами в любой точке линейной коллекции. Некоторые распространенные
примеры списков включают следующее:


*   Рецепт, представляющий собой список инструкций

*   Строка, представляющая собой список символов

*   Документ, представляющий собой список слов

*   Файл, представляющий собой список блоков данных на диске



Во всех этих примерах порядок критически важен, и перетасовка элементов делает
коллекции бессмысленными. Однако элементы в списке не обязательно сортируются. Слова
в словаре и имена в телефонной книге являются примерами отсортированных списков, но слова
в этом абзаце в равной степени образуют список и не отсортированы. Хотя элементы в списке
всегда логически смежны, они не обязательно должны быть физически смежными в памяти. Массивные
реализации списков используют физические позиции для представления логического порядка, но связанные
реализации этого не делают.

Первый элемент в списке находится в начале, в то время как последний элемент в списке находится в конце. Элементы в
списке сохраняют положение относительно друг друга с течением времени, а добавления и удаления влияют
на отношения предшественника/преемницы только в момент изменения. Специалисты по информатике
обычно подсчитывают позиции от 0 до длины списка минус 1. Каждая числовая позиция также называется индексом. При визуализации списка индексы уменьшаются влево и
увеличиваются вправо. На рисунке 9-1 показано, как список изменяется в ответ на последовательность
операции. Операции, которые представляют собой лишь небольшое подмножество тех, которые возможны для списков,
описаны в таблице 9-1.

## **Операции, основанные на местоположении**

Операции на основе содержимого основаны не на индексе, а на содержимом списка. Эти
операции обычно ожидают элемент в качестве аргумента и что-то делают с ним и списком.
Некоторые из этих операций выполняют поиск элемента, равного данному элементу, прежде чем предпринимать дальнейшие
действия. В таблице 9-3 перечислены три основные операции на основе содержимого для списка с именем L. Обратите
внимание, что для согласованности с другими коллекциями используется add вместо append.

Операции на основе позиции выполняются относительно текущей позиции, называемой
курсор. Операции позволяют программисту перемещаться по списку, перемещая этот
курсор. В некоторых языках программирования эти операции выполняет отдельный объект, называемый итератором списка
. Эта политика принята здесь. Хотя список уже поддерживает итератор,
который позволяет программисту просматривать элементы списка с помощью цикла for, итератор списка является более мощным. В отличие от простого итератора, итератор списка поддерживает перемещение к предыдущим позициям,
непосредственно к первой позиции и непосредственно к последней позиции. Помимо этих навигационных функций
операции, итератор списка также поддерживает вставки, замены и удаления элементов в
позициях курсора. Реализация итератора списка будет разработана позже в этой главе, но на данном
этапе вы можете изучить его логическую структуру и поведение.

```python
class Array(object):
    def __init__(self, size=32):
        """
                 : param Размер: длина
        """
        self._size = size
        self._items = [None] * size

         # Выполнить массив [ключ]
    def __getitem__(self, index):
        return self._items[index]

         # Выполните при массиве [ключ] = значение
    def __setitem__(self, index, value):
        self._items[index] = value

         # Выполнить, когда Лен (массив)
    def __len__(self):
        return self._size

         #   массив
    def clear(self, value=None):
        for i in range(len(self._items)):
            self._items[i] = value

         #
    def __iter__(self):
        for item in self._items:
            yield item
```

```python
class AbstractCollection(object):
    """An abstract collection implementation."""

    # Constructor
    def __init__(self, sourceCollection = None):
        """Sets the initial state of self, which includes the
        contents of sourceCollection, if it's present."""
        self._size = 0
        if sourceCollection:
            for item in sourceCollection:
                self.add(item)

    # Accessor methods
    def isEmpty(self):
        """Returns True if len(self) == 0, or False otherwise."""
        return len(self) == 0

    def __len__(self):
        """Returns the number of items in self."""
        return self._size

    def __str__(self):
        """Returns the string representation of self."""
        return "[" + ", ".join(map(str, self)) + "]"

    def __add__(self, other):
        """Returns a new bag containing the contents
        of self and other."""
        result = type(self)(self)
        for item in other:
            result.add(item)
        return result

    def __eq__(self, other):
        """Returns True if self equals other,
        or False otherwise."""
        if self is other: return True
        if type(self) != type(other) or \
           len(self) != len(other):
            return False
        otherIter = iter(other)
        for item in self:
            if item != next(otherIter):
                return False
        return True
```

```python
class AbstractList(AbstractCollection):
    """An abstract list implementation."""

    def __init__(self, sourceCollection):
        """Maintains a count of modifications to the list."""
        self.modCount = 0
        AbstractCollection.__init__(self, sourceCollection)

    def getModCount(self):
        """Returns the count of modifications to the list."""

        return self.modCount

    def incModCount(self):
        """Increments the count of modifications
        to the list."""

        self.modCount += 1

    def index(self, item):
        """Precondition: item is in the list.
        Returns the position of item.
        Raises: ValueError if the item is not in the list."""

        position = 0
        for data in self:
            if data == item:
                return position
            else:
                position += 1
        if position == len(self):
            raise ValueError(str(item) + " not in list.")

    def add(self, item):
        """Adds the item to the end of the list."""

        self.insert(len(self), item)

    def remove(self, item):
        """Precondition: item is in self.
        Raises: ValueError if item in not in self.
        Postcondition: item is removed from self."""

        position = self.index(item)
        self.pop(position)
```

```python
class ArrayListIterator(object):
    """Represents the list iterator for an array list."""

    def __init__(self, backingStore):
        """Set the initial state of the list iterator."""

        self.backingStore = backingStore
        self.modCount = backingStore.getModCount()
        self.first()

    def first(self):
        """Resets the cursor to the beginning
        of the backing store."""

        self.cursor = 0
        self.lastItemPos = -1

    def hasNext(self):
        """Returns True if the iterator has
        a next item or False otherwise."""
        return self.cursor < len(self.backingStore)

    def next(self):
        """Preconditions: hasNext returns True.
        The list has not been modified except by this
        iterator’s mutators.
        Returns the current item and advances the cursor.
        to the next item."""
        if not self.hasNext():
            raise ValueError("No next item in list iterator")
        if self.modCount != self.backingStore.getModCount():
            raise AttributeError(
                "Illegal modification of backing store")
        self.lastItemPos = self.cursor
        self.cursor += 1
        return self.backingStore[self.lastItemPos]

    def last(self):
        """Moves the cursor to the end of the backing store."""

        self.cursor = len(self.backingStore)
        self.lastItemPos = -1

    def hasPrevious(self):
        """Returns True if the iterator has a
        previous item or False otherwise."""

        return self.cursor > 0

    def previous(self):
        """Preconditions: hasPrevious returns True.
        The list has not been modified except
        by this iterator’s mutators.
        Returns the current item and moves
        the cursor to the previous item."""

        if not self.hasPrevious():
            raise ValueError("No previous item in list iterator")
        if self.modCount != self.backingStore.getModCount():
            raise AttributeError(
                "Illegal modification of backing store")
        self.cursor -= 1
        self.lastItemPos = self.cursor
        return self.backingStore[self.lastItemPos]
```

```python
class ArrayList(AbstractList):
    """An array-based list implementation."""

    DEFAULT_CAPACITY = 10

    def __init__(self, sourceCollection = None):
        """Sets the initial state of self, which includes the contents of sourceCollection, if it's present."""

        self._items = Array(ArrayList.DEFAULT_CAPACITY)
        AbstractList.__init__(self, sourceCollection)

    def _resize(self, array, logicalSize):
        """If the array needs resizing, resizes and returns the new array.  Otherwise, returns the olf array."""

        temp = None
        # If array is full
        if len(array) == logicalSize:
            temp = Array(2 * len(array))
        # If array is wasting space
        elif logicalSize <= .25 * len(array) and \
             len(array) >= ArrayList.DEFAULT_CAPACITY:
            temp = Array(round(.5 * len(array)))
        # None of the above
        else:
            return array
        # Copy items to new array
        for i in range(logicalSize):
            temp[i] = array[i]
        return temp


    # Accessor methods
    def __iter__(self):
        """Supports iteration over a view of self."""

        cursor = 0
        while cursor < len(self):
            yield self._items[cursor]
            cursor += 1

    def __getitem__(self, i):
        """Precondition: 0 <= i < len(self)
        Returns the item at position i.
        Raises: IndexError."""

        if i < 0 or i >= len(self):
            raise IndexError("List index out of range")
        return self._items[i]

    # Mutator methods
    def clear(self):
        """Makes self become empty."""

        self._size = 0
        self._items = Array(ArrayList.DEFAULT_CAPACITY)

    def __setitem__(self, i, item):
        """Precondition: 0 <= i < len(self)
        Replaces the item at position i.
        Raises: IndexError."""

        if i < 0 or i >= len(self):
            raise IndexError("List index out of range")
        self._items[i] = item

    def insert(self, i, item):
        """Inserts the item at position i."""

        self._items = self._resize(self._items, len(self))
        if i < 0: i = 0
        elif i > len(self): i = len(self)
        if i < len(self):
            for j in range(len(self), i, -1):
                 self._items[j] = self._items[j - 1]
        self._items[i] = item
        self._size += 1
        self.incModCount()

    def pop(self, i = None):
        """Precondition: 0 <= i < len(self).
        Removes and returns the item at position i.
        If i is None, i is given a default of len(self) - 1.
        Raises: IndexError."""

        if i == None: i = len(self) - 1
        if i < 0 or i >= len(self):
            raise IndexError("List index out of range")
        item = self._items[i]
        for j in range(i, len(self) - 1):
            self._items[j] = self._items[j + 1]
        self._size -= 1
        self.incModCount()
        self._items = self._resize(self._items, len(self))
        return item

    def listIterator(self):
        """Returns a list iterator."""

        return ArrayListIterator(self)

    def replace(self, item):
        """Preconditions: the current position is defined.
        The list has not been modified except by this
        iterator’s mutators."""

        if self.lastItemPos == -1:
            raise AttributeError(
                "The current position is undefined.")
        if self.modCount != self.backingStore.getModCount():
            raise AttributeError(
                "List has been modified illegally.")
        self.backingStore[self.lastItemPos] = item
        self.lastItemPos = -1

    def insert(self, item):
        """Preconditions:
        The list has not been modified except by this
        iterator’s mutators."""

        if self.modCount != self.backingStore.getModCount():
            raise AttributeError(
                "List has been modified illegally.")
        if self.lastItemPos == -1:
        # Cursor not defined, so add item to end of list
            self.backingStore.add(item)
        else:
            self.backingStore.insert(self.lastItemPos, item)
        self.lastItemPos = -1
        self.modCount += 1

    def remove(self):
        """Preconditions: the current position is defined.
        The list has not been modified except by this
        iterator’s mutators."""

        if self.lastItemPos == -1:
            raise AttributeError(
                "The current position is undefined.")
        if self.modCount != self.backingStore.getModCount():
            raise AttributeError(
                "List has been modified illegally.")
        item = self.backingStore.pop(self.lastItemPos)
        # If the item removed was obtained via next,
        # move cursor back
        if self.lastItemPos < self.cursor:
            self.cursor -= 1
        self.modCount += 1
        self.lastItemPos = -1
```

```python
print("Create a list with 1-9")
lyst = ArrayList(range(1, 10))
print("Length:", len(lyst))
print("Items (first to last): ", lyst)
# Create and use a list iterator
listIterator = lyst.listIterator()
print("Forward traversal: ", end = "")
listIterator.first()
while listIterator.hasNext():
    print (listIterator.next(), end = " ")
print("\nBackward traversal: ", end = "")
listIterator.last()
while listIterator.hasPrevious():
    print(listIterator.previous(), end = " ")
print("\nInserting 10 before 3: ", end = "")
listIterator.first()
for count in range(2):
    listIterator.next()
listIterator.insert(10)
print(lyst)
print("Removing 2: ", end = "")
listIterator.first()
for count in range(3):
    listIterator.next()
listIterator.remove()
print(lyst)
print("Removing all items")
listIterator.first()
while listIterator.hasNext():
    listIterator.next()
    listIterator.remove()
print("Length:", len(lyst))
```

### **Управление кучным хранилищем **

Когда вы читали главу 7 “Стеки”, вы узнали об одном аспекте управления памятью Python: стеке вызовов. Теперь вы познакомитесь с другими аспектами, прочитав о
том, как можно управлять свободным пространством в куче объектов, также представленном в главе 7, используя связанный список.
Напомним, что куча объектов - это область памяти, из которой виртуальная машина Python
выделяет сегменты различного размера для всех новых объектов данных. Когда на объект больше
нельзя ссылаться из программы, виртуальная машина Python (PVM) может вернуть это
объект помещает сегмент памяти в кучу для использования другими объектами. Схемы управления кучей
могут оказывать значительное влияние на общую производительность приложения, особенно если приложение создает и отбрасывает множество объектов в ходе своего выполнения.
Разработчики PVM, следовательно, готовы приложить много усилий, чтобы организовать кучу
наиболее эффективным из возможных способов. Их сложные решения выходят за рамки данной книги,
поэтому здесь представлена упрощенная схема.

В этой схеме смежные блоки свободного пространства в куче связаны в свободный список. Когда
приложение создает экземпляр нового объекта, PVM выполняет поиск в свободном списке первого блока
, достаточно большого, чтобы вместить объект. Когда объект больше не нужен, сборщик мусора
возвращает пространство объекта в список свободных.

У этой схемы есть два недостатка. Во-первых, со временем большие блоки в свободном списке фрагментируются на множество блоков меньшего размера. Во-вторых, поиск в свободном списке блоков достаточного
размера может занять O(n) времени выполнения, где n - количество блоков в списке. Чтобы противодействовать
фрагментации, сборщик мусора периодически реорганизует свободный список, объединяя соседние блоки. Чтобы сократить время поиска, вы можете использовать несколько свободных списков. Например, если
для ссылки на объект требуется 4 байта, список 1 может состоять из блоков размером 4; список 2, блоки размером
размер 8; список 3, блоки размером 16; список 4, блоки размером 32; и так далее. Последний список будет содержать
все блоки больше некоторого заданного размера.

В этой схеме пространство всегда выделяется единицами по 4 байта, а пространство для нового объекта
берется из заголовка первого непустого списка, содержащего блоки достаточного размера. Поскольку
доступ и удаление из заголовка равны O(1), выделение места для нового объекта теперь занимает O(1)
времени, если только объекту не требуется больше места, чем доступно в первом блоке последнего списка.
На этом этапе необходимо выполнить поиск по последнему списку, что дает операции максимальное время
выполнения O(n), где n - размер последнего списка.

Для простоты в этом обсуждении были полностью проигнорированы две сложные проблемы
. Первая проблема связана с принятием решения о том, когда запускать сборщик мусора.
Запуск сборщика мусора отнимает время у приложения, но его
отсутствие означает, что свободные списки никогда не пополняются. Вторая проблема касается того, как сборщик мусора идентифицирует объекты, на которые больше нет ссылок и, следовательно, они больше не
нужны. (Решение этих проблем выходит за рамки данной книги.)

### **Организация файлов на диске**

Файловая система компьютера состоит из трех основных компонентов: каталога файлов, самих файлов и свободного пространства. Чтобы понять, как они работают вместе для создания файловой системы, сначала
рассмотрим физический формат диска. На рисунке 9-4 показано стандартное расположение. Поверхность диска разделена на концентрические дорожки, и каждая дорожка далее подразделяется на секторы.
Количество этих дорожек варьируется в зависимости от емкости диска и физического размера. Однако
все дорожки содержат одинаковое количество секторов, и все секторы содержат одинаковое количество
байт. Ради этого обсуждения предположим, что сектор содержит 8 килобайт данных плюс
несколько дополнительных байт, зарезервированных для указателя. Сектор - это наименьшая единица информации, передаваемая на диск и с него, независимо от его фактического размера, и пара чисел (t, s) определяет
местоположение сектора на диске, где t - номер дорожки, а s - номер сектора. Рисунок 9-4
показывает диск с n дорожками. K секторов в дорожке 0 помечены от 0 до k – 1.

Каталог файловой системы организован как иерархическая коллекция. Здесь нет необходимости вдаваться
в подробности этой структуры. Для этих целей просто предположим, что каталог
занимает первые несколько дорожек на диске и содержит запись для каждого файла. Эта запись
содержит имя файла, дату создания, размер и так далее. Кроме того, она содержит адрес
сектора, содержащего первые байты в файле. В зависимости от его размера файл может
полностью содержаться в пределах одного сектора или может охватывать несколько секторов. Обычно, в
последний сектор заполнен только частично, и не предпринимается попыток восстановить неиспользуемое пространство.
Секторы, составляющие файл, не обязательно должны быть физически смежными, поскольку каждый сектор
, кроме последнего, заканчивается указателем на сектор, содержащий следующую часть
файла. Наконец, неиспользуемые сектора объединяются в список свободных. Когда
создаются новые файлы, им выделяется место из этого списка, а когда старые файлы удаляются, их место
возвращается в список.

Поскольку все секторы имеют одинаковый размер и поскольку пространство распределено по секторам, файловая
система не сталкивается с такой же проблемой фрагментации, как в
куче объектов Python. Тем не менее, сложность все еще существует. Чтобы перенести данные на диск или с него,
головки чтения/записи сначала должны быть установлены на правильную дорожку, диск должен вращаться
до тех пор, пока нужный сектор не окажется под головками, а затем происходит передача данных.
Из этих трех этапов передача данных занимает наименьшее время. К счастью, данные могут быть
переносятся в несколько смежных секторов или из них за один оборот без
необходимости изменения положения головок. Таким образом, производительность дисковой системы оптимизируется, когда
многосекторные файлы не разбросаны по диску. Однако со временем, по мере создания и уничтожения файлов разного размера, такого рода разброс становится частым, и
производительность файловой системы снижается. В качестве контрмеры файловые системы включают утилиту,
запускаемую либо автоматически, либо по явному запросу пользователя, которая реорганизует файл
система таким образом, чтобы секторы в каждом файле были смежными и имели одинаковый физический и
логический порядок.

### **Внедрение других коллекций**

Списки часто используются для реализации других коллекций, таких как стеки и очереди.
Есть два способа сделать это:


*   Расширьте класс list, сделав новый класс подклассом класса list.
*   Используйте экземпляр класса list внутри нового класса и позвольте списку содержать элементы данных.

Например, вы могли бы реализовать класс stack, расширив класс list. Однако расширение
в данном случае не является разумным выбором, поскольку эта версия стека наследует методы
из списка, которые позволяют пользователям получать доступ к элементам в позициях, отличных от верхней, что нарушает
дух абстрактного типа данных stack. В случае стеков и очередей лучшим дизайнерским
решением является размещение списка внутри стека или очереди. В этом случае
разработчику стека или очереди доступны все операции со списком, но только необходимый стек или очередь
операции доступны его пользователям.

Коллекции, использующие списки, также наследуют их характеристики производительности. Например, стек
, использующий список на основе массива, обладает характеристиками производительности списка на основе массива,
тогда как стек, использующий список на основе ссылок, обладает характеристиками списка на основе ссылок.

Основное преимущество использования списка для реализации другой коллекции заключается в том, что кодирование
становится проще. Вместо того, чтобы работать с конкретным массивом или связанной структурой, разработчику стека нужно только вызвать соответствующие методы списка.

### **Роль класса AbstractList**

Реализации list следуют шаблону, установленному классами bag, stack и queue
, рассмотренными в главах с 6 по 8. Вместо того, чтобы начинать с нуля, каждый конкретный класс list
является подклассом абстрактного класса, называемого AbstractList. Поскольку этот класс является подклассом
Abstract Collection, классы list наследуют обычные методы сбора, а также
переменную self.size. Можете ли вы также включить некоторые другие методы list в AbstractList, чтобы
уменьшить количество методов, которые вам нужно определить в конкретных классах?

Ответ таков: “Конечно”. Напомним, что метод index на основе содержимого - это поиск
позиции данного элемента. Поскольку для поиска может использоваться простой цикл for, вы можете определить этот
метод в абстрактном списке. Более того, методы remove и add на основе содержимого могут
вызывать методы pop и insert на основе индекса, соответственно, для удаления или добавления элемента после определения его местоположения с помощью вызова index. Таким образом, вы также можете определить эти два метода
на основе содержимого в AbstractList.

В дополнение к этим методам класс AbstractList поддерживает новую переменную экземпляра
с именем self.modCount. Итераторы списка используют эту переменную для выполнения предварительных условий для определенных методов, как обсуждается далее в этой главе. Вот код для класса AbstractList:

```python
class AbstractList(AbstractCollection):
    """An abstract list implementation."""

    def __init__(self, sourceCollection):
        """Maintains a count of modifications to the list."""
        self.modCount = 0
        AbstractCollection.__init__(self, sourceCollection)

    def getModCount(self):
        """Returns the count of modifications to the list."""

        return self.modCount

    def incModCount(self):
        """Increments the count of modifications
        to the list."""

        self.modCount += 1

    def index(self, item):
        """Precondition: item is in the list.
        Returns the position of item.
        Raises: ValueError if the item is not in the list."""

        position = 0
        for data in self:
            if data == item:
                return position
            else:
                position += 1
        if position == len(self):
            raise ValueError(str(item) + " not in list.")

    def add(self, item):
        """Adds the item to the end of the list."""

        self.insert(len(self), item)

    def remove(self, item):
        """Precondition: item is in self.
        Raises: ValueError if item in not in self.
        Postcondition: item is removed from self."""

        position = self.index(item)
        self.pop(position)
```

### **Реализация на основе массива**

Реализацией интерфейса list на основе массива является класс, называемый ArrayList.
ArrayList хранит свои элементы данных в экземпляре класса Array, представленного в
Главе 4 “Массивы и связанные структуры”. ArrayList имеет начальную емкость по умолчанию, которая
автоматически увеличивается при необходимости

Поскольку класс ArrayList является подклассом AbstractList, здесь определены только методы на основе индекса, метод
__iter__ и методы ListIterator.

Операции на основе индекса __getitem__ и __setitem__ просто используют оператор подстрочного индекса для переменной массива self.items. Методы insert и pop перемещают элементы в
массиве, используя методы, описанные в главе 4. Обсуждение класса-итератора ArrayList
отложено до более позднего раздела этой главы. Вот код для класса ArrayList:

```python
class ArrayList(AbstractList):
    """An array-based list implementation."""

    DEFAULT_CAPACITY = 10

    def __init__(self, sourceCollection = None):
        """Sets the initial state of self, which includes the contents of sourceCollection, if it's present."""

        self._items = Array(ArrayList.DEFAULT_CAPACITY)
        AbstractList.__init__(self, sourceCollection)

    def _resize(self, array, logicalSize):
        """If the array needs resizing, resizes and returns the new array.  Otherwise, returns the olf array."""

        temp = None
        # If array is full
        if len(array) == logicalSize:
            temp = Array(2 * len(array))
        # If array is wasting space
        elif logicalSize <= .25 * len(array) and \
             len(array) >= ArrayList.DEFAULT_CAPACITY:
            temp = Array(round(.5 * len(array)))
        # None of the above
        else:
            return array
        # Copy items to new array
        for i in range(logicalSize):
            temp[i] = array[i]
        return temp


    # Accessor methods
    def __iter__(self):
        """Supports iteration over a view of self."""

        cursor = 0
        while cursor < len(self):
            yield self._items[cursor]
            cursor += 1

    def __getitem__(self, i):
        """Precondition: 0 <= i < len(self)
        Returns the item at position i.
        Raises: IndexError."""

        if i < 0 or i >= len(self):
            raise IndexError("List index out of range")
        return self._items[i]

    # Mutator methods
    def clear(self):
        """Makes self become empty."""

        self._size = 0
        self._items = Array(ArrayList.DEFAULT_CAPACITY)

    def __setitem__(self, i, item):
        """Precondition: 0 <= i < len(self)
        Replaces the item at position i.
        Raises: IndexError."""

        if i < 0 or i >= len(self):
            raise IndexError("List index out of range")
        self._items[i] = item

    def insert(self, i, item):
        """Inserts the item at position i."""

        self._items = self._resize(self._items, len(self))
        if i < 0: i = 0
        elif i > len(self): i = len(self)
        if i < len(self):
            for j in range(len(self), i, -1):
                 self._items[j] = self._items[j - 1]
        self._items[i] = item
        self._size += 1
        self.incModCount()

    def pop(self, i = None):
        """Precondition: 0 <= i < len(self).
        Removes and returns the item at position i.
        If i is None, i is given a default of len(self) - 1.
        Raises: IndexError."""

        if i == None: i = len(self) - 1
        if i < 0 or i >= len(self):
            raise IndexError("List index out of range")
        item = self._items[i]
        for j in range(i, len(self) - 1):
            self._items[j] = self._items[j + 1]
        self._size -= 1
        self.incModCount()
        self._items = self._resize(self._items, len(self))
        return item

    def listIterator(self):
        """Returns a list iterator."""

        return ArrayListIterator(self)

    def replace(self, item):
        """Preconditions: the current position is defined.
        The list has not been modified except by this
        iterator’s mutators."""

        if self.lastItemPos == -1:
            raise AttributeError(
                "The current position is undefined.")
        if self.modCount != self.backingStore.getModCount():
            raise AttributeError(
                "List has been modified illegally.")
        self.backingStore[self.lastItemPos] = item
        self.lastItemPos = -1

    def insert(self, item):
        """Preconditions:
        The list has not been modified except by this
        iterator’s mutators."""

        if self.modCount != self.backingStore.getModCount():
            raise AttributeError(
                "List has been modified illegally.")
        if self.lastItemPos == -1:
        # Cursor not defined, so add item to end of list
            self.backingStore.add(item)
        else:
            self.backingStore.insert(self.lastItemPos, item)
        self.lastItemPos = -1
        self.modCount += 1

    def remove(self):
        """Preconditions: the current position is defined.
        The list has not been modified except by this
        iterator’s mutators."""

        if self.lastItemPos == -1:
            raise AttributeError(
                "The current position is undefined.")
        if self.modCount != self.backingStore.getModCount():
            raise AttributeError(
                "List has been modified illegally.")
        item = self.backingStore.pop(self.lastItemPos)
        # If the item removed was obtained via next,
        # move cursor back
        if self.lastItemPos < self.cursor:
            self.cursor -= 1
        self.modCount += 1
        self.lastItemPos = -1
```

### **Связанная реализация**

Ранее в этой книге были использованы связанные структуры для реализации пакетов, стеков и очередей.
Структура, используемая для связанного стека (см. главу 7), которая имеет указатель на начало, но не
на конец, была бы неразумным выбором для связанного списка. Метод добавления списка должен был бы пройти по
цепочке через всю последовательность узлов, чтобы найти конец списка.

Односвязная структура, используемая для связанной очереди (см. главу 8, “Очереди”), работала
бы лучше, поскольку сохраняется указатель как на конец структуры, так и на ее начало.
Метод добавления списка помещает новый элемент в конец связанной структуры и при необходимости корректирует головную ссылку
.

Однако для удаления элемента в последней позиции в односвязной структуре все равно
потребуется перейти по ссылкам от начала структуры вниз к узлу, предшествующему последнему
. Таким образом, извлечение последнего элемента, которое является операцией постоянного времени для списка на основе массива,
становится операцией линейного времени для связанного списка. Более того, односвязная структура не
является идеальной для поддержки итератора списка. Итератор списка допускает перемещение курсора в любом
направлении, но односвязная структура поддерживает перемещение только к следующему узлу. Ты можешь
решите эти проблемы, используя двусвязную структуру, где каждый узел имеет указатель на
предыдущий узел, а также указатель на следующий узел.

Код, необходимый для управления двусвязной структурой, может быть упрощен, если добавить один дополнительный
узел в начале структуры, как упоминалось в главе 4. Этот узел называется
контрольным узлом или фиктивным узлом заголовка, и он указывает вперед на первый узел данных и
назад на последний узел данных. Внешний указатель заголовка указывает на контрольный узел.
Результирующая структура напоминает структуру с круговой связью, представленную в главе 4. Узел
sentinel не содержит элемента списка, и когда список пуст, остается sentinel.
На рисунке 9-5 показана пустая циклически связанная структура и циклически связанная структура, содержащая один элемент данных.

Как вы можете видеть из рисунка, следующий указатель сторожевого узла определяет местоположение первого узла данных,
тогда как его предыдущий указатель определяет местоположение последнего узла данных. Таким образом, в реализации нет необходимости в отдельном
хвостовом указателе. Более того, как вы вскоре увидите, когда вставляется или удаляется первый или последний
узел данных, нет необходимости сбрасывать головной
указатель реализации.

Основным строительным блоком двусвязной структуры является узел с двумя указателями: next,
который указывает вправо; и previous, который указывает влево. Этот тип узла, называемый TwoWayNode,
является подклассом класса Node, определенного в главе 4.

Следующий сегмент кода показывает установочный код для класса LinkedList, а также его метод
__iter__:

```python
class LinkedList(object):

    def __init__(self, head=None):
        self.head = head

    def __len__(self):
        """ Return the length of the linkedlist, not including the head node. """
        length = 0
        if self.head is None or self.head.next is None:
            return None
        node = self.head.next
        while node:
            length += 1
            node = node.next
        return length

    def __contains__(self, item):
        """ To suggest whether the item is in the linkedlist or not. """
        if self.head is None or self.head.next is None:
            return None
        node = self.head
        while node:
            if node is item or node.data == item:
                return True
            node = node.next
        return False

    def __reversed__(self):
        """ Return the reversed linkedlist. e.g. head->node1->node2->node3 ==> node3->node2->node1->head. Now node3 is the now head node. """
        if self.head is None or self.head.next is None:
            return None
        curl = self.head
        prev = None
        # pntr = None
        while curl:
            # pntr = curl
            temp = curl.next
            curl.next = prev
            prev = curl
            curl = temp
        # return pntr
        return prev

    def isEmpty(self):
        return len(self) == 0

    def append(self, value):
        """ Append a new node after the tail of the linkedlist. Special case: if head node is none, let the new item be the head. """
        item = self.Node(value)
        if self.head is None:
            self.head = item
        else:
            node = self.head
            while node.next:
                node = node.next
            node.next = item

    def butt(self, value, old_head_value):
        """ Put  a new node before the head node of the linkedlist. Special case: if head node is none, let the new item be the head."""
        item = self.Node(value)
        if self.head is None:
            self.head = item
        else:
            self.head.data = old_head_value     # Give a data field to the old head node.
            item.next = self.head
            self.head = item



    def printNode(self):
        """ Traversal the linkedlist, and print the data of the nodes. """
        if self.head is None or self.head.next is None:
            return None
        node = self.head.next
        while node is not None:
            print(node.data, end=" ")
            node = node.next


    def reverse(self):
        """ Reverse the linkedlist. e.g. head->node1->node2->node3 ==> node3->node2->node1->head. Now node3 is the now head node. """
        if self.head is None or self.head.next is None:
            return None
        prev = None
        curl = self.head
        # pntr = None
        while curl:
            # pntr = curl
            temp  = curl.next
            curl.next = prev
            prev = curl
            curl = temp
        # self.head = pntr
        self.head = prev



    class Node(object):

        def __init__(self, data=None, next=None):
            self.data = data
            self.next = next
```

Обратите внимание, что метод __init__ создает узел без данных — это сторожевой узел.
Затем метод __iter__ устанавливает свой курсор не на головной узел, который является сторожевым, а на
следующий узел, который является первым узлом, содержащим данные, если этот узел существует. Когда курсор
перемещается к головному узлу, цикл итератора завершается.

Еще предстоит разработать методы на основе индексов __getitem__, __setitem__,
insert и pop. Каждый из этих методов должен проходить по цепочке через узлы в связанной
структуре, начиная с узла после головного узла, пока не будет достигнут i-й узел. В
этот момент данные, содержащиеся в этом узле, возвращаются или изменяются (__getitem__ или
__setitem__), или узел удаляется (pop), или новый узел вставляется перед этим узлом
(insert). Поскольку поиск i-го узла - это операция, которую должны выполнять все четыре метода
для выполнения вы включаете вспомогательный метод с именем getNode, который выполняет это. Этот метод
ожидает позицию индекса целевого узла в качестве аргумента и возвращает указатель на i-й узел. Затем четыре вызывающих метода могут использовать этот указатель для соответствующего манипулирования
связанной структурой.

Стоит обратить внимание на использование метода getNode как в __set item__, так и в insert. Каждый из этих
методов проверяет, находится ли индекс элемента в диапазоне, прежде чем вызывать getNode для определения местоположения
узла. Кроме того, поскольку связанная структура включает в себя сторожевой узел, метод insert
не должен обрабатывать особые случаи вставок в начале и конце

Как упоминалось ранее, итератор списка - это объект, присоединенный к списку, который обеспечивает позиционные
операции над этим списком. Эти операции, перечисленные в таблицах 9-4 и 9-5, позволяют программисту

списка на основе массива; версия на основе ссылок оставлена для вас в качестве упражнения.

### **Роль и обязанности итератора списка**

Когда программист запускает метод ListIterator для списка, этот метод возвращает новый
экземпляр класса list iterator. Объект итератора списка зависит от связанного списка в том смысле, что
первому необходимо получить доступ ко второму, чтобы найти элементы, заменить их, вставить или удалить
. Таким образом, итератор списка будет поддерживать ссылку на свой список или резервное хранилище, которое он
получает при его создании.

Чтобы помочь определить некоторые из этих предварительных условий, итератор списка поддерживает две дополнительные
переменные. Первая - это его собственная переменная mod count. Этой переменной присваивается значение
mod count списка при создании итератора списка. Таким образом, каждый список и итератор списка имеют
свое собственное “представление” о количестве модификаций. Всякий раз, когда запускается собственный мутатор списка, он увеличивает
количество модификаций списка, чтобы записать модификацию в список. Когда определенные методы, такие как next
и previous, запускаются в итераторе списка, итератор списка сравнивает свой собственный modcount с
количество модификаций в списке. Если два значения различны, кто-то запустил модификатор списка в
неправильном контексте; тогда возникает исключение. Когда метод list iterator изменяет список,
list iterator увеличивает свое собственное количество модификаций, чтобы поддерживать согласованность двух значений формы.

Вторая переменная отслеживает позицию, в которой итератор списка может выполнить мутацию в
списке. В реализации на основе массива эта переменная равна -1 всякий раз, когда позиция еще
не установлена. Его значение становится индексом в списке всякий раз, когда программист
успешно выполняет следующую или предыдущую операцию в итераторе списка. Таким образом, методы-мутаторы insert и remove в итераторе списка могут проверить эту переменную на наличие их предварительного условия и
сбросить ее до -1 после успешного изменения списка.

## **Настройка и создание экземпляра класса итератора списка**

Напомним, что метод ArrayList ListIterator передает резервное хранилище (self)
итератору списка во время его создания. Затем итератор списка может запускать методы списка для этого
объекта, чтобы манипулировать им. Вот код для этой части класса ArrayListIterator.

```python
class ArrayListIterator(object):
    """Represents the list iterator for an array list."""

    def __init__(self, backingStore):
        """Set the initial state of the list iterator."""

        self.backingStore = backingStore
        self.modCount = backingStore.getModCount()
        self.first()

    def first(self):
        """Resets the cursor to the beginning
        of the backing store."""

        self.cursor = 0
        self.lastItemPos = -1

    def hasNext(self):
        """Returns True if the iterator has
        a next item or False otherwise."""
        return self.cursor < len(self.backingStore)

    def next(self):
        """Preconditions: hasNext returns True.
        The list has not been modified except by this
        iterator’s mutators.
        Returns the current item and advances the cursor.
        to the next item."""
        if not self.hasNext():
            raise ValueError("No next item in list iterator")
        if self.modCount != self.backingStore.getModCount():
            raise AttributeError(
                "Illegal modification of backing store")
        self.lastItemPos = self.cursor
        self.cursor += 1
        return self.backingStore[self.lastItemPos]

    def last(self):
        """Moves the cursor to the end of the backing store."""

        self.cursor = len(self.backingStore)
        self.lastItemPos = -1

    def hasPrevious(self):
        """Returns True if the iterator has a
        previous item or False otherwise."""

        return self.cursor > 0

    def previous(self):
        """Preconditions: hasPrevious returns True.
        The list has not been modified except
        by this iterator’s mutators.
        Returns the current item and moves
        the cursor to the previous item."""

        if not self.hasPrevious():
            raise ValueError("No previous item in list iterator")
        if self.modCount != self.backingStore.getModCount():
            raise AttributeError(
                "Illegal modification of backing store")
        self.cursor -= 1
        self.lastItemPos = self.cursor
        return self.backingStore[self.lastItemPos]
```

## **Рекурсивный обход списка, подобного Lisp**

Учитывая рекурсивное определение списка, подобного Lisp, и его основные операции доступа,
теперь мы можем определить некоторые рекурсивные функции, которые просматривают списки. Функция содержит поиск в
списке заданного элемента. Эта функция ожидает целевой элемент и список в качестве аргументов и возвращает
True или False. Если список пуст, у вас закончились элементы для проверки, поэтому функция
возвращает False. В противном случае, если целевой элемент равен первому элементу в списке, функция
возвращает True. В противном случае вы используете contains для рекурсивного поиска в остальной части списка в поисках
данный элемент. Вот код для этой функции:

```python
def contains(item, lyst):
    """Returns True if item is in lyst or
    False otherwise."""

    if isEmpty(lyst):
        return False
    elif item == first(lyst):
        return True
    else:
        return contains(item, rest(lyst))
```

Функция get, основанная на индексе, возвращает i-й элемент данного списка. Предполагается, что аргумент
index находится в диапазоне от 0 до длины аргумента list минус 1. Функция
, по сути, продвигается по списку и ведет обратный отсчет от заданного индекса до 0. Когда
индекс достигает 0, функция возвращает первый элемент в списке на данный момент. Каждый рекурсивный
вызов не только уменьшает индекс, но и переходит к остальной части списка. Определение
get следует:

```python
def get(index, lyst):
    """Returns the item at position index in lyst.
    Precondition: 0 <= index < length(lyst)"""

    if index == 0:
        return first(lyst)
    else:
        return get(index - 1, rest(lyst))
```

Предположим, вы не знаете длину списка, которая является частью предварительного
условия функции get. Определение длины списка, подобного Lisp, может быть задано рекурсивно. Его длина
равна 0, если список пуст. В противном случае длина списка равна единице плюс длина остальной части списка
после первого элемента. Вот код для длины рекурсивной функции:

```python
def length(lyst):
    """Returns the number of items in lyst."""

    if isEmpty(lyst):
        return 0
    else:
        return 1 + length(rest(lyst))
```

Наиболее важным в этих обходах является то, что они следуют рекурсивной структуре
списка. Широкий спектр функций рекурсивной обработки списков может быть определен просто в терминах
базовых функций доступа к списку isEmpty, first и rest.

## **Создание списка, похожего на Lisp**

Теперь можно узнать, как создать список, подобный Lisp. Список, подобный Lisp, имеет единственную базовую функцию-конструктор с именем cons. Эта функция ожидает два аргумента: элемент данных и другой
список. Функция создает и возвращает новый список, первый элемент которого является первым аргументом функции. Остальные элементы в новом списке содержатся во втором аргументе функции.

Давайте используем эту информацию для определения рекурсивной функции, которая возвращает список, содержащий
диапазон последовательных чисел. Границы этого диапазона являются аргументами нашей функции,
которую вы называете build Range. Например, вызов build Range(1, 5) возвращает список
, содержащий 1, 2, 3, 4 и 5, а build Range(10, 10) возвращает список, содержащий только 10.
Обобщая, если границы равны, build Range возвращает список, содержащий одну из них. В противном случае build Range возвращает список, первый элемент которого является нижней границей, а остальные
элементы содержат список, построенный из диапазона между нижней границей плюс единица и верхней
границей. Вот код для построения диапазона, за которым следует объяснение:

```python
def buildRange(lower, upper):
    """Returns a list containing the numbers from
    lower through upper.
    Precondition: lower <= upper"""

    if lower == upper:
        return cons(lower, THE_EMPTY_LIST)
    else:
        return cons(lower, buildRange(lower + 1, upper))
```

Рекурсивный шаблон в только что рассмотренной функции встречается во многих других
функциях обработки списков. В качестве еще одного примера рассмотрим проблему удаления элемента в i-й
позиции списка. Если эта позиция является первой (0), то мы возвращаем остальную часть списка.
В противном случае мы возвращаем список, построенный из первого элемента, и список, который получается в результате удаления
элемента в этой позиции минус единица из остальной части списка. Как и функция get, рассмотренная
ранее, функция remove уменьшает индекс и переходит к остальной части списка при каждом рекурсивном вызове.
Вот код:

```python
def remove(index, lyst):
    """Returns a list with the item at index removed.
    Precondition: 0 <= index < length(lyst)"""

    if index == 0:
        return rest(lyst)
    else:
        return cons(first(lyst),
            remove(index - 1, rest(lyst)))
```

## **Внутренняя структура Lisp-подобного списка**

Как показано на рисунке 9-10, внутренняя структура списка, подобного Lisp
, напоминает односвязную структуру, представленную в главе 4. Эта структура состоит из последовательности узлов,
где каждый узел содержит элемент данных с именем data и ссылку на следующий узел с именем next.
Следующая ссылка в последнем узле - None. Если вы определяете символ THE_EMPTY_LIST как None,
то вы можете использовать класс Node из главы 4 для представления узлов в Lisp-подобном списке.
Определения четырех основных функций списка тривиальны, как показано в следующем фрагменте кода:

```python
THE_EMPTY_LIST = None

def isEmpty(lyst):
    """Returns True if lyst is empty or False otherwise."""
    return lyst is THE_EMPTY_LIST

def first(lyst):
    """Returns the item at the head of lyst.
    Precondition: lyst is not empty."""

    return lyst.data

def rest(lyst):
    """Returns a list of the items after the head of lyst.
    Precondition: lyst is not empty."""

    return lyst.next

def cons(data, lyst):
    """Returns a list whose head is item and whose tail
    is lyst."""

    return Node(data, lyst)
```

Важно помнить, что список, подобный Lisp, представляет собой абстрактный тип данных (ADT), который
включает в себя эти четыре основные функции и константу для пустого списка. Пользователь этого
ADT не должен ничего знать об узлах, ссылках или указателях.

## **Печать списков, похожих на Lisp, в режиме ожидания с помощью __repr__**

Списки Lisp имеют буквальное представление, когда они печатаются в интерпретаторе Lisp. Это
представление аналогично представлению списков Python, но в нем используются круглые скобки вместо скобок
и пропущены запятые между элементами. Пустой список печатается как специальный символ, такой как nil.

Чтобы выполнить это, вы добавляете метод с именем __repr__ в класс Node. Когда Python
вычисляет значение для отображения в оболочке IDLE, он ищет этот метод в классе значения. Если
вы включите этот метод в класс Node, вы можете создать и вернуть строку с данными
всей последовательности узлов, заключенными в круглые скобки. Вот код для класса Node, который
включает этот метод:

```python
class Node(object):
    """Represents a singly linked node."""

    def __init__(self, data, next = None):

        self.data = data
        self.next = next

    def __repr__(self):
        """Returns the string representation of a nonempty lisp list."""

        def buildString(lyst):

            if isEmpty(rest(lyst)):
                return str(first(lyst))
            else:
                return str(first(lyst)) + " " + buildString(rest(lyst))

        return "(" + buildString(self) + ")"
```

Обратите внимание, что метод __repr__ использует вложенную функцию с именем buildstring для рекурсии
по связанной структуре. Поскольку рекурсивная стратегия использует функции isEmpty,
first и rest, теперь вы включаете определение класса Node в модуль lisp list
.

## **Списки и функциональное программирование**

Одна из интересных особенностей Lisp-подобных списков, по крайней мере, в том виде, в каком мы их определили, заключается в том, что
в них нет операций-мутаторов. Даже функция remove, разработанная ранее, не
изменяет структуру своего аргумента list; она просто возвращает список с удаленным i-м элементом.

Если бы в этих списках были операции мутатора, такое совместное использование структуры было бы плохой
идеей, потому что любые изменения в структуре списка A затем привели бы к изменениям в
структуре списка B. Однако, когда никакие мутации невозможны, совместное использование структуры является хорошей
идеей, поскольку это может сэкономить память.

Списки, подобные Lisp, без мутаторов прекрасно вписываются в стиль разработки программного обеспечения, называемый
функциональным программированием. Программа, написанная в этом стиле, состоит из набора взаимодействующих
функций, которые преобразуют значения данных в другие значения данных. Таким образом, когда структура данных
должна быть изменена, она не мутируется, а вместо этого передается в качестве аргумента функции.
Функция создает и возвращает структуру данных, которая представляет желаемые изменения.

Преимущество этого стиля программирования заключается в том, что можно легко проверить правильность работы функций
, поскольку все изменения, которые они вносят в данные, прозрачны (никаких скрытых
побочных эффектов).

С другой стороны, затраты времени выполнения на запрещение мутаций могут быть дорогостоящими.
Например, метод удаления на основе индекса, рассмотренный ранее в этой главе, не требует дополнительной
памяти для удаления элемента из списка. Напротив, удаление элемента в позиции i
из списка, подобного Lisp, требует i – 1 дополнительных узлов.

Эти компромиссы привели к старой шутке о том, что программисты на Lisp знают ценность
всего и стоимость ничего. Очевидно, что объектно-ориентированная структура данных, поддерживающая
мутацию, была бы лучшим выбором для приложений с большими базами данных, которые требуют частых
вставок и удалений. Но для обработки относительно коротких списков символьной информации
существует немного структур данных, столь простых и элегантных, как рекурсивный список, подобный Lisp.
