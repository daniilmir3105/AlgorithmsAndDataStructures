#**Стэки**

Стопки представляют собой линейные коллекции, в которых доступ полностью ограничен только одним концом, называемым
верхним. Классическим аналогичным примером является стопка чистых подносов, которую можно найти в каждом кафетерии.
Всякий раз, когда требуется поднос, он снимается с верхней части стопки, и всякий раз, когда чистые
тарелки возвращаются с кухни, они снова ставятся сверху. Никто никогда не берет
какой-нибудь особенно мелкий лоток из середины стопки, а лотки, расположенные ближе к дну, могут
вообще не использоваться. Говорят, что стопки придерживаются протокола LIFO ("последний вход - первый выход").

Операции по укладке предметов в стопку и извлечению предметов из стопки называются push и
pop соответственно. На рисунке 7-1 показана стопка в том виде, в каком она может выглядеть на разных этапах. Элемент в
верхней части стопки заштрихован.

Другие повседневные примеры стопок включают тарелки и миски в кухонном шкафу или стопку компакт-дисков.

## **Создание стэков**

Тип стека не встроен в Python. В крайнем случае, программисты на Python могут использовать
список на Python для эмуляции стека на основе массива. Если вы рассматриваете конец списка как вершину стека, метод
list append помещает элемент в этот стек, тогда как метод list pop
удаляет и возвращает элемент в его верхней части. Основным недостатком этой опции является то, что все
остальные операции со списком также могут манипулировать стеком. К ним относятся вставка,
замена и удаление элемента в любой позиции. Эти дополнительные операции нарушают
дух стека как абстрактного типа данных. В этом разделе определяется более ограниченный интерфейс
для любой аутентичной реализации стека и показано, как эти операции используются в кратком
примере.

## **Интерфейс стэка**

В дополнение к операциям push и pop, интерфейс стека предоставляет операцию с именем
peek для проверки элемента в верхней части стека. Как и другие коллекции, тип stack
также может включать операции clear, isEmpty, len, str, in и +, а также итератор.
Эти операции перечислены как методы Python в таблице 7-1, где переменные ссылаются на

Обратите внимание, что методы pop и peek имеют важное предварительное условие и вызывают исключение, если пользователь стека не удовлетворяет этому предварительному условию. Преимущество этого интерфейса в том, что пользователи будут знать, какие методы использовать и чего от них ожидать, независимо

### **Вычисление арифметических выражений**

В повседневной жизни люди настолько привыкли вычислять простые арифметические выражения, что
мало задумываются о соответствующих правилах. Итак, вы можете быть удивлены сложностью написания алгоритма для вычисления арифметических выражений. Оказывается, косвенный подход
к проблеме работает лучше всего. Сначала вы преобразуете выражение из его привычной инфиксной формы
в постфиксную форму, а затем вычисляете постфиксную форму. В инфиксной форме каждый оператор
расположен между своими операндами, тогда как в постфиксной форме оператор следует непосредственно за
его операнды. В таблице 7-4 приведены несколько простых примеров.

Между этими двумя формами есть сходства и различия. В обеих операнды отображаются в
одинаковом порядке. Однако операторы этого не делают. Инфиксная форма иногда требует круглых скобок; постфиксная форма никогда этого не делает. Вычисление инфикса включает правила приоритета; постфиксное вычисление применяет операторы, как только они встречаются. Например, рассмотрим этапы
вычисления инфиксного выражения 34 + 1 22 * 2 и эквивалентного постфиксного выражения 34 22 2 * +.

*   Связывание переменных с объектами данных, хранящимися в памяти, чтобы их можно было найти при

## **Реализация массива**

Первая реализация построена вокруг массива self.items и целого числа
self.size. Изначально емкость массива по умолчанию составляет 10 позиций, а self.size. равно 0.
Верхний элемент, если он есть, всегда будет находиться в местоположении self.size – 1. Чтобы поместить элемент
в стек, сохраните элемент в местоположении self.items[len(self)] и увеличьте
self.size. Чтобы заполнить стек, верните self.items[len(self) - 1] и уменьшите self.size.
На рисунке 7-6 показано, как отображаются self.items и self.size, когда в стеке четыре элемента.

```python
class AbstractStack():
    """An abstract stack implementation."""

    # Constructor
    def __init__(self, sourceCollection):
        """Sets the initial state of self, which includes the
        contents of sourceCollection, if it's present."""

        AbstractCollection.__init__(self, sourceCollection)

    # Mutator methods
    def add(self, item):
        """Adds item to self."""

        self.push(item)
```

Вот код для Array Stack, с некоторыми частями, которые необходимо выполнить в упражнениях:

```python
class ArrayStack(AbstractStack):
    """An array-based stack implementation."""

    DEFAULT_CAPACITY = 10 # For all array stacks

    def __init__(self, sourceCollection = None):
        """Sets the initial state of self, which includes the
        contents of sourceCollection, if it's present."""

        self.items = Array(ArrayStack.DEFAULT_CAPACITY)
        AbstractStack.__init__(self, sourceCollection)

    # Accessors
    def __iter__(self):
        """Supports iteration over a view of self.
        Visits items from bottom to top of stack."""

        cursor = 0
        while cursor < len(self):
            yield self.items[cursor]
            cursor += 1

    def peek(self):
        """Returns the item at top of the stack.
        Precondition: the stack is not empty.
        Raises KeyError if the stack is empty."""

        # Check precondition here
        return self.items[len(self) - 1]

    # Mutators
    def clear(self):
        """Makes self become empty."""

        self.size = 0
        self.items = Array(ArrayStack.DEFAULT_CAPACITY)

    def push(self, item):
        """Inserts item at top of the stack."""

        # Resize array here if necessary
        self.items[len(self)] = item
        self.size += 1

    def pop(self):
        """Removes and returns the item at top of the stack.
        Precondition: the stack is not empty.
        Raises KeyError if the stack is empty.
        Postcondition: the top item is removed from the stack."""

        # Check precondition here
        oldItem = self.items[len(self) - 1]
        self.size -= 1

        # Resize the array here if necessary
        return oldItem
```

## **Связанная реализация**

Как и реализация linked bag из главы 6, связанная реализация стека использует
односвязную последовательность узлов. Эффективное перемещение и выскакивание требует добавления и удаления узлов в начале связанной последовательности. Переменная экземпляра self.items теперь
ссылается на узел в начале этой последовательности, если таковой имеется. В противном случае, когда стек
пуст, self.items равно None.

Хотя связанная структура поддерживает простое нажатие и pop, реализация метода
__iter__ усложняется тем фактом, что элементы должны быть посещены из хвоста
связанной структуры в ее начало. К сожалению, чтобы пройти односвязную структуру, нужно
 начать с ее начала и перейти по следующим ссылкам к ее концу.

К счастью, рекурсия может прийти на помощь. В методе __iter__ вы создаете временный список и определяете рекурсивную вспомогательную функцию, которая ожидает узел в качестве аргумента. На
при первоначальном вызове функции узел аргумента является главой связанной структуры стека (
переменная self.items). Если этот узел не равен None, вы вызываете функцию рекурсивно со
следующим полем узла, чтобы перейти к концу структуры. Когда этот вызов возвращается,
вы добавляете данные узла во временный список. Когда возвращается вызов вспомогательной
функции верхнего уровня, вы возвращаете итератор в списке.

Вот код для LinkedStack:

```python
class LinkedStack(AbstractStack):
    """ Link-based stack implementation."""

    def __init__(self, sourceCollection = None):

        self.items = None
        AbstractStack.__init__(self, sourceCollection)

    # Accessors
    def __iter__(self):
        """Supports iteration over a view of self.
        Visits items from bottom to top of stack."""

        def visitNodes(node):
            if node != None:
                visitNodes(node.next)
                tempList.append(node.data)
        tempList = list()
        visitNodes(self.items)
        return iter(tempList)

    def peek(self):
        """Returns the item at top of the stack.
        Precondition: the stack is not empty."""

        if self.isEmpty():
            raise KeyError("The stack is empty.")
        return self.items.data

    # Mutators
    def clear(self):
        """Makes self become empty."""

        self.size = 0
        self.items = None

    def push(self, item):
        """Inserts item at top of the stack."""

        self.items = Node(item, self.items)
        self.size += 1

    def pop(self):
        """Removes and returns the item at top of the stack.
        Precondition: the stack is not empty."""

        if self.isEmpty():
            raise KeyError("The stack is empty.")
        oldItem = self.items.data
        self.items = self.items.next
        self.size -= 1
        return oldItem
```

## **Временной и пространственный анализ двух реализаций**

За исключением метода __iter__, все методы стека просты и имеют
максимальное время выполнения O(1), или константу. При реализации массива анализ становится более сложным. В момент удвоения время выполнения метода push равно
переходит к (n) или линейному, но в остальное время оно остается равным O(1). Аналогичные замечания можно
сделать о методе pop. В среднем оба значения по-прежнему равны O(1), как показано в главе 4. Однако программист должен решить, приемлемо ли колеблющееся время отклика, и соответствующим образом выбрать
реализацию.

Метод __iter__ выполняется за линейное время в обеих реализациях. Однако рекурсивная
функция, используемая в связанной реализации, вызывает линейный рост объема памяти из
-за использования стека системных вызовов. Вы можете избежать этой проблемы, используя двусвязную структуру; итератор может начинаться с последнего узла и переходить по ссылкам на предыдущие узлы. Глава 9,
“Списки”, подробно рассматривает такую структуру.

Коллекции из n объектов требуется, по крайней мере, достаточно места для хранения n ссылок на объекты. Давайте
теперь посмотрим, как две наши реализации стека соотносятся с этим идеалом. Связанный стек из n
элементов требует n узлов, каждый из которых содержит две ссылки: одну на элемент, а другую на
следующий узел. Кроме того, должна быть переменная, указывающая на верхний узел, и переменная
для размера, что дает общее требование к пространству, равное 2n + 2.

Для реализации массива общее требование к пространству стека фиксируется, когда стек равен
создан экземпляр. Пространство состоит из массива с начальной емкостью 10 ссылок и
переменных для отслеживания размера стека и ссылок на сам массив. Предполагая, что целое число
и ссылка занимают одинаковый объем пространства, общее требование к пространству
равно емкости массива + 2. Как обсуждалось в главе 4, реализация массива более экономична с точки зрения пространства
, чем связанная реализация, когда коэффициент загрузки превышает ½. Коэффициент загрузки
для реализации массива обычно варьируется от ¼ до 1, хотя, очевидно, он может
опуститься до 0.
