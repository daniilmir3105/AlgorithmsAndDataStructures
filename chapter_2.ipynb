# **Обзор коллекций в питоне **

Как известно, Python включает в себя несколько встроенных типов коллекций: string, список,
кортеж, набор и словарь. Строка и список, вероятно, являются наиболее распространенными и фундаментальными типами коллекций. Другие важные типы коллекций включают
стеки, очереди, очереди с приоритетом, деревья бинарного поиска, кучи, графики, пакеты и различные типы
отсортированных коллекций. Коллекции могут быть однородными, что означает, что все элементы в коллекции должны быть одного типа, или разнородными, что означает, что элементы могут быть разных типов. Во многих языках программирования коллекции однородны, хотя большинство
Коллекций Python могут содержать несколько типов объектов.

Коллекции обычно являются динамическими, а не статическими, что означает, что они могут увеличиваться или уменьшаться
в зависимости от потребностей задачи. Кроме того, их содержимое может изменяться в ходе выполнения
программы. Одним из исключений из этого правила является неизменяемая коллекция, такая как строка
или кортеж Python. Элементы неизменяемой коллекции добавляются во время ее создания; после этого никакие элементы
не могут быть добавлены, удалены или заменены.

Другой важной отличительной характеристикой коллекций является способ, которым они
организованы. Теперь мы рассмотрим организацию, используемую в нескольких широких категориях коллекций: линейные коллекции, иерархические коллекции, графические коллекции, неупорядоченные коллекции
и отсортированные коллекции.

## **Линейные коллекции**

Элементы линейной коллекции, как и люди в очереди, упорядочены по позиции. У каждого элемента
, кроме первого, есть уникальный предшественник, и у каждого элемента, кроме последнего, есть уникальный
преемник. Как показано на рисунке 2-1, предшественником D2 является D1, а преемником D2 является D3.

## **Иерархические коллекции**

Элементы данных в иерархических коллекциях упорядочены в структуре, напоминающей перевернутое дерево. У каждого элемента данных, кроме того, что находится вверху, есть только один предшественник, называемый его
родительским, но потенциально много преемников, называемых его дочерними элементами. Как показано на рисунке 2-2,
предшественником (родительским элементом) D3 является D1, а преемниками (дочерними элементами) D3 являются D4, D5 и D6.

## **Коллекции графов**

Коллекция графов, также называемая графом, представляет собой коллекцию, в которой у каждого элемента данных может быть много
предшественников и много преемников. Как показано на рисунке 2-3, все элементы, подключенные к D3
, считаются как его предшественниками, так и его преемниками, и они также называются его соседями.

Примерами графиков являются карты маршрутов авиакомпаний между городами, схемы электропроводки
зданий и всемирная паутина.

## **Неупорядоченные коллекции**

Как следует из названия, элементы в неупорядоченной коллекции не расположены в каком-либо определенном порядке, и
невозможно осмысленно говорить о предшественнике или преемнике элемента. На рисунке 2-4
показана такая структура.

Снимок экрана 2024-03-12 163906.png

Пакет с шариками - пример неупорядоченной коллекции. Хотя вы можете класть шарики
в пакет и вынимать их из пакета в любом порядке, в пакете шарики
расположены в произвольном порядке.

Отсортированная коллекция накладывает естественный порядок на ее элементы. Примерами могут служить записи в
телефонной книге (разновидность бумажной книги 20-го века) и имена в списке классов.

Чтобы установить естественный порядок, должно быть какое-то правило для сравнения элементов, например
item(i) <= item(i+1)
, для элементов, посещенных в отсортированной коллекции.

Хотя отсортированный список является наиболее распространенным примером отсортированной коллекции, отсортированные коллекции
необязательно быть линейным или упорядоченным по позиции. С точки зрения клиента, наборы, сумки и словари могут быть отсортированы, даже если их элементы недоступны по позиции. Особый тип иерархической коллекции, известный как бинарное дерево поиска, также устанавливает естественный порядок для своих элементов.

Отсортированная коллекция позволяет клиенту просматривать все ее элементы в отсортированном порядке. Некоторые операции,
такие как поиск, могут быть более эффективными для отсортированной коллекции, чем для ее несортированного собрата.

## **Преобразование типов**


Одна из операций сбора данных, с которой вы, возможно, не знакомы, - это преобразование типов. Вы уже
знаете о преобразовании типов по его использованию при вводе чисел. В этом контексте вы преобразуете строку цифр с клавиатуры в int или float, применяя функцию int или float
к входной строке.

Можно преобразовать один тип коллекции в другой тип коллекции аналогичным образом.
Например, вы можете преобразовать строку Python в список Python, а список Python
в кортеж Python, как показано в следующем сеансе:

```python
message = "Hi there!"
lyst = list(message)
lyst
```

```python
toople = tuple(lyst)
toople
```

## **Клонирование и равенство**

Особым случаем преобразования типов является клонирование, которое возвращает точную копию аргумента
функции преобразования. Это должно быть в том случае, когда тип аргумента совпадает
с типом функции преобразования. Например, следующий сегмент кода создает копию списка, а
затем сравнивает два списка, используя операторы is и ==. Поскольку два списка не являются одним и тем
же объектом, is возвращает значение False. Поскольку два списка являются различными объектами, но относятся к одному
типу и имеют одинаковую структуру (каждая пара элементов одинакова в каждой позиции в
двух списках), == возвращает True.

```python
lyst1 = [2, 4, 8]
lyst2 = list(lyst1)
lyst1 is lyst2
```

```python
lyst1 == lyst2
```
